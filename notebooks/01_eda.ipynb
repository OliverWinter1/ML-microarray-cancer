```python
import pandas as pd
import GEOparse

gse = GEOparse.get_GEO("GSE199633", destdir="data")
df = gse.pivot_samples("VALUE")

print(df.shape)
```

Using pyarrow allows us to save the data to a parquet file, allow faster processing of large-data sets

```python
import pyarrow

X = df.T
X.to_parquet("data/X.parquet")
print(X.shape)
```
