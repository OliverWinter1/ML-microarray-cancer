Read the datasets and make sure that X and y is both defined and aligned 

```python
import pandas as pd

X = pd.read_parquet("data/X.parquet")
y = pd.read_csv("data/y.csv")

y.rename(columns={'Unnamed: 0': 'Sample',
    'characteristics_ch1.4.er status': 'Er_status'}, inplace=True)
X.rename(columns={'Unnamed: 0': 'Sample'},inplace=True)

y.set_index('Sample', inplace=True)

print(X.shape)
print(y.shape)

assert X.index.equals(y.index)
```

Bar Chart to compare the distribution of Er+ and Er-. 

```python
import matplotlib.pyplot as plt

counts = y["Er_status"].value_counts().sort_index()
plt.bar(counts.astype(str), counts.values)
plt.xlabel("Er Status")
plt.ylabel("Number")
plt.title("Distribution of Estrogen receptor status between samples of breast cancer tumours")
plt.xticks([0, 1], ["Negative", "Positive"])
plt.grid(axis="y", color="lightgray", alpha=0.3)
```
